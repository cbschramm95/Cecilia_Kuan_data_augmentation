# ===DATA WITH POSITIVE or POS+NEG SAMPLES ===

## ADD LABELS FOR DATA SET ""EXCEPT NEGATIVE DATASET""

```python
# df1_neg.to_pickle('/extradisk/data/cecilia_k/a-proof-zonmw/data/jenia/train_eb_ap_jenia_neg-only.pkl')
# df1_pos.to_pickle('/extradisk/data/cecilia_k/a-proof-zonmw/data/jenia/train_eb_ap_jenia_pos-only.pkl')

inpkl = '../data/sharona/sharona_all_data.pkl'
outpkl = '../data/sharona/sharona_all_labels.pkl'

df1 = pd.read_pickle(inpkl)
df1
```

```python

# ADD BINARY GOLD LABELS 'labels_binary'
df1['labels_binary'] = df1['labels'].apply(lambda x: 0 if x == [0] * len(x) else 1) #return 0

# ADD 10-CAT GOLD LABEL, 'labels_10'
df1['labels_10'] = df1['labels'].apply(lambda x: x + [1] if x == [0] * len(x) else x + [0])

df1 = df1.rename({'labels':'labels_9', }, axis = 1)
print(df1)

# SAVE PKL WITH ALL LABEL FORMATS (9 CATEGORIES(ORIGINAL), 10 CATEGORIES, BINARY)
df1.to_pickle(outpkl)
```
