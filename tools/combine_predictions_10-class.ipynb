# Combine predictions from 2 classifiers for evaluation:
- Update multi-label predictions from 9 to 10-class format;
- Final pkl is for calculating performance with ***10-class*** format - 9 x ICF + 1 x "None"

```python
import pandas as pd
import numpy as np
```

```python
# INPUT & OUTPUT FILE PATHS

binary_neg = '../data/output/b2_NEG_output_sharona_data_all_labels.pkl'
icf_pred = '../data/output/e5_output_b2_POS_sharona_data_all_labels.pkl'

combined_outfile = '../data/output/combined_output_b2_e5_sharona_data_all_labels.pkl'
```

```python
# LOAD DATA TO DATAFRAMES
df_bneg = pd.read_pickle(binary_neg)
df_icf = pd.read_pickle(icf_pred)
```

### BINARY classifier Predictions

```python
# neg ex was not predicted by ICF classifier, set to NaN
df_bneg['pred_9']= df_bneg['year'].apply(lambda x: [0] * 9)

# prediction - update prediction with 10-cat format
df_bneg['pred_10'] = df_bneg['year'].apply(lambda x: [0] * 9 + [1])

df_bneg
```

### MULTI-LABEL classifier predictions

```python
df_icf= df_icf.rename({'pred_jenia':'pred_9'}, axis = 1)

# prediction - update prediction with 10-cat format
df_icf['pred_10'] = df_icf['pred_9'].apply(lambda x: x + [1] if x == [0] * len(x) else x + [0])

df_icf
```

```python
# COMBINE PREDICTIONS FOR EVALUATION

combined_df = pd.concat([df_icf,df_bneg], ignore_index=True)
combined_df.to_pickle(combined_outfile)
```
