# Separate BINARY CLASSIFIER output

- separate predictions of "0" and "1"
- Input: .pkl with binary classifier predictions
- Output: POS.pkl with "1" predictions for prediction process with ICF classifier
- Output: NEG.pkl with "0" predictions

```python
import pandas as pd
import random
```

```python
binary_output_inpkl ='../data/output/b2_output_train_eb_ap_jenia_all-labels.pkl'
neg_output_outpkl = '../data/output/b2_NEG_output_train_eb_ap_jenia_all-labels.pkl'
pos_output_outpkl = '../data/output/b2_POS_output_train_eb_ap_jenia_all-labels.pkl'

df = pd.read_pickle(binary_output_inpkl)
df
```

```python
df_neg = df.loc[df['pred_binary']== 0]
df_pos = df.loc[df['pred_binary']== 1]

df_neg.to_pickle(neg_output_outpkl)
df_pos.to_pickle(pos_output_outpkl)
```
